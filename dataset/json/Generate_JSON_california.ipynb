```python
import pandas as pd
import json
```

```python
df_california = pd.read_csv(
    '../california-quality-of-life.csv'
)
```

```python
df_california = df_california.drop(columns=['Category'])
```

```python
df_california.columns.tolist()
```

```python
df_california_sorted = df_california.sort_values(by=['Year', 'Sex', 'Question', 'Disease'])
df_california_sorted = df_california_sorted[[
    'Year',
    'Sex',
    'Question',
    'Disease',
    'Data_Value',
    'Sample_Size',
    'Low_Confidence_Limit',
    'High_Confidence_Limit',
    'Count',
    'Population',
    'Rate',
    'CI.lower',
    'CI.upper'
]]
df_california_sorted
```

```python
# df_california_sorted.drop(df_california_sorted[df_california_sorted['Disease'] != "Chlamydia"].index, inplace = True)
```

```python
# group by
df_california_groupby_SexYearQuestion = df_california_sorted.groupby(by=["Sex","Year","Question"], dropna=False).mean()

# make index into columns
df_california_groupby_SexYearQuestion = df_california_groupby_SexYearQuestion.reset_index(level=2).reset_index(level=1).reset_index(level=0)

df_california_groupby_SexYearQuestion
```

```python
# store file to react

def write_dict_to_file_California(name, dict_data):
    with open("../../react-app/src/data/california_"+name+".json", "w") as write_file:
        json.dump(dict_data, write_file, indent=2)
```

```python
# parse df to json

def parse_df_to_json_records(data):
    json_data = data.to_json(orient="records")
    return json.loads(json_data)

# search in list of dict

def find(lst, key, value):
    for i, dic in enumerate(lst):
        if dic[key] == value:
            return i
    return -1
```

```python
# unflatten original dict

def parse_carlifornia_groupby_YearSexQuestion(data):
    
    dict_data = {}
    
    for d in data:
        dict_data[d['Sex']] = []
        
    for d in data:
        sex_index = find(dict_data[d['Sex']], 'Year', d['Year'])
        
        if (sex_index == -1):
            dict_data[d['Sex']].append({
                'Year': d['Year'],
                 d['Question'].replace(" ", "_"): {
                    
                    'Data_Value': d['Data_Value'],
                    'Sample_Size': d['Sample_Size'],
                    'Low_Confidence_Limit': d['Low_Confidence_Limit'],
                    'High_Confidence_Limit': d['High_Confidence_Limit'],

                    'Count': d['Count'],
                    'Population': d['Population'],
                    'Rate': d['Rate'],
                    'CI.lower': d['CI.lower'],
                    'CI.upper': d['CI.upper']
                     
                }
            })
        else:
            dict_data[d['Sex']][sex_index][d['Question'].replace(" ", "_")] = {
                
                'Data_Value': d['Data_Value'],
                'Sample_Size': d['Sample_Size'],
                'Low_Confidence_Limit': d['Low_Confidence_Limit'],
                'High_Confidence_Limit': d['High_Confidence_Limit'],
                
                'Count': d['Count'],
                'Population': d['Population'],
                'Rate': d['Rate'],
                'CI.lower': d['CI.lower'],
                'CI.upper': d['CI.upper']
            }
            

    print(json.dumps(dict_data, indent=2))
    return dict_data
```

```python
parsed_california_data = parse_df_to_json_records(df_california_groupby_SexYearQuestion)
dict_california_data = parse_carlifornia_groupby_YearSexQuestion(parsed_california_data)
write_dict_to_file_California("groupby_SexYear", dict_california_data)
```
