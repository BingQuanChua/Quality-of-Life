## Mode of Transmission

```python
mode = pd.read_csv('dataset/disease-mode-of-transmission.csv')
mode.head()
```

```python
mode_count = {}
for index, row in mode.iterrows():
    modes = [x.strip() for x in row['Mode_of_Transmission'].split(',')]
    for m in modes:
        if mode_count.get(m) is None:
            mode_count[m] = 1
        else:
            mode_count[m] = mode_count.get(m)+1
    print(f'"{row["Disease"]}": {{"mode": {modes}}},')
```

```python
mode_count
```

```python
mode_count = {k: v for k, v in sorted(mode_count.items(), key=lambda item: item[1])}
mode_count
```

```python
import matplotlib.pyplot as plt
```

```python
# Data to plot
labels = []
sizes = []

for x, y in mode_count.items():
    labels.append(f"{x}: {y}")
    sizes.append(y)

# Plot
plt.pie(sizes, labels=labels)

plt.axis('equal')
plt.show()
```

## Merging the Two

```python
quality_of_life.head()
```

```python
quality_of_life.Year.unique()
```

```python
quality_of_life = quality_of_life[quality_of_life['Year'].notna()]
```

```python
quality_of_life['Year'] = quality_of_life['Year'].astype(int)
```

```python
quality_of_life.info()
```

```python
quality_of_life
```

```python
infectious_disease

infectious_disease.rename(columns = {'County':'LocationDesc'}, inplace=True)
```

```python
infectious_disease
```

```python
infectious_disease['Sex'] =  [x.replace('Total', 'Overall') for x in infectious_disease['Sex']]
```

```python
infectious_disease
```

```python
california_quality_of_life = pd.merge(quality_of_life, 
                                      infectious_disease,
                                      left_on  = ['Year', 'LocationDesc', 'Break_Out'], 
                                      right_on = ['Year', 'LocationDesc', 'Sex'])
```

```python
california_quality_of_life = california_quality_of_life[california_quality_of_life['LocationDesc']=='California']
california_quality_of_life
```

```python
to_drop = ['LocationAbbr', 'DataSource', 'Data_Value_Unit', 
           'Data_Value_Footnote_Symbol', 'Data_Value_Footnote', 'Data_Value_Std_Err',
           'GeoLocation', 'CategoryId', 'TopicId', 'QuestionId', 'LocationId',
           'BreakOutId', 'BreakOutCategoryid', 'Unstable']
```

```python
california_quality_of_life['Year'].value_counts()
```

```python
california_quality_of_life['LocationDesc'].value_counts()
```

```python
msno.matrix(california_quality_of_life)
```

```python
california_quality_of_life = california_quality_of_life.drop(to_drop, axis=1)
california_quality_of_life.reset_index(inplace=True, drop=True)
```

```python
to_drop2 = [
    'Break_Out_Category',
    'Break_Out',
    'Topic', 'LocationDesc', 'Data_Value_Type'
]

california_quality_of_life = california_quality_of_life.drop(to_drop2, axis=1)
california_quality_of_life.reset_index(inplace=True, drop=True)
```
